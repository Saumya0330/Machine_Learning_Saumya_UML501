(d) Find the most frequent value and their indice(s) in the following arrays:


i.
x = np.array([1,2,3,4,5,1,2,1,1,1])


ii.
y = np.array([1, 1, 1, 2, 3, 4, 2, 4, 3, 3, ])

```python
x = np.array([1,2,3,4,5,1,2,1,1,1])

def calc(x):
    print("Array:", x)

    freq = np.bincount(x)
    max_freq = freq.max()

    most_freq_vals = np.where(freq == max_freq)[0]

    print("Most frequent value(s):", most_freq_vals)
    print("Frequency:", max_freq)

    for val in most_freq_vals:
        indices = np.where(x == val)[0]
        print(f"Indices for {val}:", indices)

y = np.array([1,1,1,2,3,4,2,4,3,3])
calc(x)
print()
calc(y)
```

(f) For the matrix: n_array = np.array([[55, 25, 15],[30, 44, 2],[11, 45, 77]]), find

i.
Sum of diagonal elements

ii. Eigen values of matrix

iii. Eigen vectors of matrix

iv. Inverse of matrix

v. Determinant of matrix

```python
n = np.array([[55,25,15],[30,44,2],[11,4,77]])
print('sum of diagonal elements: ',np.trace(n))
eig_values, eig_vectors = np.linalg.eig(n)
print('eigen values: ',eig_values)
print('eigen vectors: ',eig_vectors)
print('inverse: ',np.linalg.inv(n))
print('determinant: ',np.linalg.det(n))
```

(g) Multiply the following matrices and also find covariance between matrices using NumPy:

i.
p = [[1, 2], [2, 3]]
q = [[4, 5], [6, 7]]

ii.
p = [[1, 2], [2, 3], [4, 5]]
q = [[4, 5, 1], [6, 7, 2]]

```python
def mult_cov(m,n):
  print(np.matmul(m,n))
  print(np.cov(m.flatten() ,n.flatten()))
  #print(np.cov(n))

p = np.array([[1,2],[2,3]])
q = np.array([[4,5],[6,7]])
mult_cov(p,q)

p1 = np.array([[1,2],[2,3],[4,5]])
q1 = np.array([[4,5,1],[6,7,2]])
mult_cov(p1,q1)
```

#Q2: Based on NumPy Mathematics and Statistics

(a)
For the array: array = np.array([[1, -2, 3],[-4, 5, -6]])

i. Find element-wise absolute value

ii. Find the 25th, 50th, and 75th percentile of flattened array, for each column, for each
row.

iii. Mean, Median and Standard Deviation of flattened array, of each column, and
each row


```python
a = np.array([[1,-2,3],[-4,5,-6]])
print(np.absolute(a))

# Percentile of flattened array
print(np.percentile(a.flatten(), [25,50,75]))
# Percentile for each column (axis=0)
print(np.percentile(a, [25,50,75], axis=0))
# Percentile for each row (axis=1)
print(np.percentile(a, [25,50,75], axis=1))

print(np.mean(a.flatten()))
print(np.mean(a, axis=0))
print(np.mean(a, axis=1))

print(np.median(a.flatten()))
print(np.median(a, axis=0))
print(np.median(a, axis=1))

print(np.std(a.flatten()))
print(np.std(a, axis=0))
print(np.std(a, axis=1))
```

(a) Write a function named img_to_array(path) that reads an image from a specified path and
save it as text file on local machine? (Note: use separate cases for RGB and Grey Scale
images)

```python
import cv2
import os

def img_to_array(path):
  img = cv2.imread(path, cv2.IMREAD_UNCHANGED)
  if img is None:
    print('Image not found')
    return
  base = os.path.splitext(os.path.basename(path))[0]

  if len(img.shape) ==2:
    print("Image is Greyscale")
    output_path = f'{base}_greyscale.txt'
    np.savetxt(output_path, img, fmt='%d')
    print(f"Grayscale image array saved to {output_path}")
  elif len(img.shape )==3 and img.shape[2]==3:
    print("Image is RGB")
    output_path = f"{base}_rgb.txt"
    h, w, _ = img.shape
    reshaped = img.reshape(-1, 3)
    np.savetxt(output_path, reshaped, fmt="%d")
    print(f"RGB image array saved to {output_path}")
  else:
    print("Unsupported image format.")

img_to_array('/content/saumya_donut_.jpg')
```
